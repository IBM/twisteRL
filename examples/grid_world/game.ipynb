```python
from twisterl.train import prepare_algorithm, load_config
```

## Prepare algorithm and train

```python
config = load_config("ppo_grid_world_5x5_v1.json")
algo = prepare_algorithm(config)
```

```python
algo.learn(num_steps=350)
```

#### Helper function to display env

```python
import numpy as np

def display(state):
    txt = ""
    for row in np.array(state).reshape([config["env"]["width"], config["env"]["height"]]):
        txt += "|" + "|".join(("{:d}").format(num) for num in row) + "|\n"
    print(txt)
```

## Play!

#### Generate an initial state

```python
algo.env.reset()
state0 = algo.env.get_state()
print("start:")
display(algo.env.get_state())
```

#### Solve it

```python
solution = algo.solve(algo.env.get_state(), deterministic=False, num_searches=10)
len(solution)
```

#### Replay the solution step by step

```python
algo.env.set_state(state0)

print("start:")
display(algo.env.get_state())
for a in solution:
    print(["up", "down", "right", "left"][a]+":")
    algo.env.step(a)
    display(algo.env.get_state())
print(f"Solution steps: {len(solution)}")
```
